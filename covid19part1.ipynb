```python
import numpy as np
import pandas as pd
import io
import requests
import matplotlib.pyplot as plt
```

```python
url="https://api.covid19india.org/csv/latest/state_wise.csv"
s=requests.get(url).content
```

```python
df = pd.read_csv(io.StringIO(s.decode('utf-8')))
```

```python
df.info()
```

```python
df.head()
```

```python
plt.figure(figsize=(12,4))
plt.bar(x=df['State'],height=df['Active'],edgecolor='k',color='orange')
plt.grid(True)
plt.xticks(rotation=90,fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.bar(x=df['State'],height=df['Confirmed'],edgecolor='k',color='orange')
plt.grid(True)
plt.xticks(rotation=90,fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```
