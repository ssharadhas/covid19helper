```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = 'https://api.covid19india.org/csv/latest/district_wise.csv'
df = pd.read_csv(path)
df.info()
df.head()
```

```python
df.drop([ 'State_Code','District_Key',], axis=1, inplace=True)
df.rename(columns={'District': "District Name"}, inplace=True)
df.head()
```

```python
### group the data by district

world = df.groupby("District Name")['Confirmed','Active','Recovered','Deceased'].sum().reset_index()
world.head()
```

```python
### Find top 20 district with maximum number of confirmed cases
top_20 = world.sort_values(by=['Active'], ascending=False).head(20)
### Generate a Barplot
plt.figure(figsize=(12,10))
plot = sns.barplot(top_20['Active'], top_20['District Name'])
for i,(value,name) in enumerate(zip(top_20['Active'],top_20['District Name'])):
  plot.text(value,i-0.05,f'{value:,.0f}',size=10)
plt.show()
```

```python
top_5 = world.sort_values(by=['Active'], ascending=False).head()

### Generate a Barplot
plt.figure(figsize=(15,5))
active = sns.barplot(top_5['Active'], top_5['District Name'], color = 'red', label='Active')

### Adding Texts for barplots

for i,(value,name) in enumerate(zip(top_5['Active'],top_5['District Name'])):
  active.text(value,i-0.05,f'{value:,.0f}',size=9)

plt.legend(loc=4)
plt.show()
```
